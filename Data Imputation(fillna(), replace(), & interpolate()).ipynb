###Data Imputation: How to put/impute data using Python 

```python
import pandas as pd
from matplotlib import axes
import numpy as np
```

```python
#import a csv file first
data=pd.read_csv("/content/time_series_covid19_confirmed_global.csv")
df=pd.DataFrame({'A':[12,3,2,1,'not applicable'],
              'B':[2,3,2,np.nan,None],
              'C':[23,1,2,1,"NaN"],
              'D':[21,'NaN','np.nan',12,12],
              'E':[1,2,None,12,12]})
df.to_csv("/content/time_series_covid19_confirmed_global.csv")
print(df)
```
