```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import axes
import pandas as pd

fig, ax= plt.subplots()
iris = sns.load_dataset('iris')
ax.hist(iris['petal_length'], bins = 20)
ax.set_title("Histogram")
ax.set_xlabel("hah")
ax.set_ylabel("wah")
```

```python
#Seaborn Histogram
sns.distplot(iris['petal_length'], bins=10, kde= False)
```
