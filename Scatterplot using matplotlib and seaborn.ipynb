```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import axes
import pandas as pd
```

```python
iris=sns.load_dataset("iris")
#create a figure and axis
fig, ax= plt.subplots()
ax.scatter(iris['sepal_length'],iris['petal_width'])

#title and label
ax.set_title("ScatterPlot")
ax.set_xlabel("Sepal length")
ax.set_ylabel("Petal width")
```

```python
# print(iris, iris.iloc[90][4])
#plot a figure and axis
fig, ax=plt.subplots()

#make a dict of colors
colors={"setosa":'r',"versicolor":'g',"virginica":'b'}

#plot scatterplot
for i in range(len(iris["sepal_length"])):
  ax.scatter(iris["sepal_length"][i], iris["sepal_width"][i], color=colors[iris["species"][i]])

#giving title and lebels
ax.set_title("Scatterplot")
ax.set_xlabel("SepalLength")
ax.set_ylabel("SepalWidth")
```

```python
#Seaborn Scatterplot:
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris)
```

```python
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
ax.set_title("Scatterplot using seaborn")
```

```python
sns.regplot(x="sepal_length", y="sepal_width", data=iris)
```

```python
sns.lmplot(x="sepal_length",y="sepal_width", hue="species", data=iris)
```

```python
sns.swarmplot(x="species", y="petal_length", data=iris)
```
