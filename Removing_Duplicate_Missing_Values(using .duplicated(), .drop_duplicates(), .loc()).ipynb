```python
data=pd.read_csv("/content/vgsalesGlobale.csv")
print(data)
```

```python
#let's check for duplicated values in a particular column
data.Rank.duplicated()  #return bool values
```

```python
data.duplicated()
```

```python
data.duplicated().sum() #checking for sum of all the duplicated value
```

```python
#if I just need to check for duplicated values in some bunch of columns then:
data.duplicated(subset=['Rank', 'Platform', 'Genre','Name']).sum()
```

```python
data[data.duplicated(subset=['Name', 'Genre', 'Platform', 'Rank'])] #for the rows which are duplicated
```

```python
#return all of the rows and columns which do not have any duplicated values in it.
data[data.duplicated(subset=['Rank', 'Platform', 'Genre','Name'])==False]
```

```python
#Now I am creating  some duplicated values samples 
sample1=data.sample(5)
data2=data.append(sample1)
```

```python
print(data2)
#let's now check for the duplicated samples
print("\n\n------Duplicated sample rows are:------\n\n")
data2[data2.duplicated()]
```

```python
#for the sum of the duplicated values
data.duplicated().sum()
data2.duplicated().sum()
```

```python
# data2.loc[9819,]
lst1=[]
for i in range(0,5):
  lst1.append(data2.index[data2.duplicated()][i])
data2.loc[lst1,]
```

```python
data2.drop_duplicates()
```

```python
#we have two same rows so it is just dropping first row and keeping the second one
data2.drop_duplicates(keep='last').reset_index() 
```
