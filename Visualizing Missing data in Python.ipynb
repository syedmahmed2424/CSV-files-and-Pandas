```python
#you need a csv file to store df in it.
import pandas as pd
import numpy as np
import seaborn as sns
df=pd.DataFrame({'A':[12,234,6,"N/a",9],
                 'B':[1,2,5,6,None],
                 'C':[4,5,6,3,'N/A'],
                 'D':[12,4,6,"n/a", "NaN"],
                 'E':[np.nan, "not applicable",12,3,3]})
df.to_csv("/content/time_series_covid19_confirmed_global.csv")


print(df)


df=pd.read_csv("/content/time_series_covid19_confirmed_global.csv", index_col=[0], na_values=["N/A","N/a","not applicable"])

print(df.head())
```

```python
#using heat map:#A heatmap is a plot of rectangular data as a color-encoded matrix
sns.heatmap(df.isnull(), yticklabels=False, annot= True)
```
